```python
df_no_target = df[df['target1'] == ' ']
df = df[df['target1'] != ' '] # remove rows with target1 as ' '
```

```python
#pd.value_counts(df['target1'].values, sort=False)
df.target1.unique() # values in target var are STR -> INT
```

```python
df['target1'] = df['target1'].map({'0': 0, '1': 1})
```

```python
df.info() # now all columns are int64
```

```python
#missing data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(5)
```

```python
# Let's start 
# Return item and drop from frame. Raise KeyError if not found.
y = df.pop('target1') # target lebel
X = df # observations
```

```python
from sklearn.model_selection import train_test_split
# performing 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(len(X_train), len(X_test))
```

```python
# Model 1 : Decision Tree
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=7, min_samples_split=4,)
clf.fit(X_train, y_train)
```

```python
# Training accuracy
print('Training Accuracy: '+ str(clf.score(X_train, y_train)))
# Cross Validation: model's accuracy using 10 fold cross validation
from sklearn.model_selection import cross_val_score
print('CV training accuracy ' + str(np.mean(cross_val_score(clf, X_train, y_train,cv=10, n_jobs=-1, verbose=10))))
```

```python
# Test Predictions
y_pred = clf.predict(X_test)
print(clf.score(X_test,y_test))
```
